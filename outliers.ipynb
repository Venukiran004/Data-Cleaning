```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
my_data=pd.read_csv("BankCreditCard.csv")
```

```python
my_data=my_data.drop("Customer ID",1)
```

```python
plt.figure(figsize=(15,15)) 
plt.style.use('ggplot')
sns.boxplot(data = my_data,width=.8,orient="h") 
```

```python
lower=[]
for i in col_data:
    Q1=my_data[i].quantile(0.25)
    Q3=my_data[i].quantile(0.75) 
    lower_bound=Q1-(1.5*(Q3-Q1))
    lower.append(lower_bound)  
lower  
```

```python
upper=[]
for i in col_data:
    Q1=my_data[i].quantile(0.25)
    Q3=my_data[i].quantile(0.75)
    upper_bound=Q3+(1.5*(Q3-Q1))
    upper.append(upper_bound)  
upper  
```

```python
for i,j,k in zip(lower,upper,col_data):
    my_data=my_data[(my_data[k]>=i) & (my_data[k]<=j)]
    
```
